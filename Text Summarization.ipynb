```python
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
import json
import re
```

```python
url = "https://www.bbc.com/news/world-europe-66601553"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

article = soup.find("article")

text = article.get_text()

tokens = word_tokenize(text)

num_tokens = len(tokens)
data_length = len(text)
```

```python
print("\033[93m Number of Tokens:",'\033[91m', num_tokens)
print("\033[93m Length of Scrapped Data:",'\033[91m', data_length)
```

```python
print(text)
```

```python
cleaned_text = re.sub('\s+', ' ', text).strip()
```

```python
def split_text(text):
    max_chunk_size = 1800
    chunks = []
    current_chunk = ""
    
    # Tokenize the text into sentences
    sentences = sent_tokenize(text)
    
    for sentence in sentences:
        if len(current_chunk) + len(sentence) < max_chunk_size:
            current_chunk += sentence + " "
        else:
            chunks.append(current_chunk.strip())
            current_chunk = sentence + " "
    
    if current_chunk:
        chunks.append(current_chunk.strip())
    
    return chunks

chunks = split_text(cleaned_text)
chunks
```

```python
import openai

openai.api_key = api_key

def generate_summary(chunk):
    response = openai.Completion.create(
        engine="davinci",
        prompt=(f"Please summarize the following text:\n{chunk}\n\nSummary:"),
        temperature=0.5,
        max_tokens=1024,
        n=1,
        stop=None
    )
    summary = response.choices[0].text.strip()
    return summary

summaries = [generate_summary(chunk) for chunk in chunks]

for i, summary in enumerate(summaries, start=1):
    print(f"Summary for Chunk {i}:\n{summary}\n")
```
